```python
# get the correlation matrix
corr_matrix = df_train.corr()

# Filter the correlation matrix to include only the desired values
filtered_corr_matrix = corr_matrix[(corr_matrix >= 0.3) | (corr_matrix <= -0.3)]

# plotting the correlation matrix
plt.figure(figsize = (30,30))
sns.heatmap(filtered_corr_matrix, annot = True)
plt.show()
```

```python
# Set the correlation threshold
threshold = 0.3

# Iterate over the correlation matrix
for i in range(len(corr_matrix.columns)):
    for j in range(i+1, len(corr_matrix.columns)):
        correlation = corr_matrix.iloc[i, j]
        
        # Check if correlation exceeds the threshold
        if abs(correlation) >= threshold:
            column_1 = corr_matrix.columns[i]
            column_2 = corr_matrix.columns[j]
            
            # Print the correlated column pair
            print(f"{column_1} and {column_2} are correlated with correlation {correlation:.2f}")
```

```python
# Set the correlation threshold
threshold = 0.3

# Initialize a list to store the selected column names
selected_columns = []

# Iterate over the correlation matrix
for i in range(len(corr_matrix.columns)):
    for j in range(i+1, len(corr_matrix.columns)):
        correlation = corr_matrix.iloc[i, j]
        
        # Check if correlation exceeds the threshold
        if abs(correlation) >= threshold:
            column_1 = corr_matrix.columns[i]
            column_2 = corr_matrix.columns[j]
            
            # Add the correlated column names to the selected columns list
            selected_columns.extend([column_1, column_2])

# Remove duplicate column names and create a new dataframe with the selected columns
filtered_df = df_train[selected_columns].copy()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

def scatterplot_matrix(df, column_pairs):
    """
    Plot scatterplots for specified pairs of columns in a dataframe using Seaborn.
    
    Parameters:
    - df: pandas DataFrame
    - column_pairs: list of lists, each sublist contains two column names
    
    Returns:
    - None (displays the scatterplot matrix)
    """
    sns.set(style='ticks')
    for pair in column_pairs:
        sns.scatterplot(data=df, x=pair[0], y=pair[1])
        plt.show()
```

```python
x = filtered_df.drop(columns = 'Class')
y = filtered_df[['Class']]
Id = filtered_df[['Id']]
```

```python
# spliting the data in to test and train sets
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.15)
```

```python
x_train.drop(['Id'], axis= 1, inplace=True)
test_ID = x_test[['Id']]
x_test.drop(['Id'], axis= 1, inplace=True)
```

```python
from sklearn.utils import shuffle, resample

# Combine the feature matrix X and target labels y
data = np.concatenate((x_train, y_train), axis=1)

# Separate the majority and minority classes
majority_class = data[data[:, -1] == 0]
minority_class = data[data[:, -1] == 1]

# Upsample the minority class to match the majority class size
minority_upsampled = resample(minority_class,
                              replace=True,   # Sample with replacement
                              n_samples=len(majority_class),   # Match majority class size
                              random_state=42)   # Set random state for reproducibility

# Combine the upsampled minority class with the majority class
data_balanced = np.concatenate((majority_class, minority_upsampled))

# Shuffle the balanced data
data_balanced = shuffle(data_balanced, random_state=42)

# Separate the feature matrix X_balanced and target labels y_balanced
X_balanced = data_balanced[:, :-1]
y_balanced = data_balanced[:, -1]
```

```python
# scaling the data before feeding the model
from sklearn.preprocessing import StandardScaler, MinMaxScaler

scaler_x = StandardScaler()
x = scaler_x.fit_transform(X_balanced)
x_test = scaler_x.fit_transform(x_test)
#scaler_y = StandardScaler()
#y = scaler_y.fit_transform(target)
```

```python
# create a random forest model
from sklearn.ensemble import RandomForestClassifier
import xgboost as xgb

rf = RandomForestClassifier(random_state = 42)
rf.fit(x_train, y_train)

xgb_clf = xgb.XGBClassifier()
xgb_clf.fit(x_train, y_train)
```

```python
# predict for the test dataset
y_predict = rf.predict(x_test)

y_predict_1 = xgb_clf.predict(x_test)
```
